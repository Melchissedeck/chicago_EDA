```python
import pandas as pd

# -- Apercu rapide du fichier reviews.csv --
# Adapter le chemin selon l'emplacement du fichier sur ta machine
reviews = pd.read_csv("reviews.csv")

print("=== Shape ===")
print(reviews.shape)

print("\n=== Colonnes ===")
print(reviews.columns.tolist())

print("\n=== Types ===")
print(reviews.dtypes)

print("\n=== Premieres lignes ===")
print(reviews.head(5))

print("\n=== Valeurs manquantes ===")
print(reviews.isnull().sum())

print("\n=== Statistiques de base ===")
print(reviews.describe(include="all"))

print("\n=== Plage temporelle ===")
if "date" in reviews.columns:
    reviews["date"] = pd.to_datetime(reviews["date"])
    print(f"Date min : {reviews['date'].min()}")
    print(f"Date max : {reviews['date'].max()}")
    print(f"Nombre de mois couverts : {reviews['date'].dt.to_period('M').nunique()}")
```
